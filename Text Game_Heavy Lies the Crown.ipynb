```python
# Title of the game : Heavy Lies the Crown. 

"""
Docstring: 
A) This game is based on my love for fantasy fiction both in English and 
my native language Tamil. The game puts forth few scenarios where a king
must make decisions without much information.
Room 1: Small introduction 
Room 2: Choice between Battle and Consulting the Oracle
Room 3: The Temple
Room 4: Opening the Snake Chamber
Some unknown terms you might face:
Yazhi: https://en.wikipedia.org/wiki/Yali_(mythology)
Rudraveena: https://en.wikipedia.org/wiki/Rudra_veena
Mantra: is a chant or a slogan.

B) Since the story aspect is heavy, the player might get confused while scrolling
down. I have added '---' lines to make the story more readable. Not aware of anyother bugs.
"""
# Creating a colour pallette
#Colours
red = '\033[31m'
black = '\033[30m'
yellow = '\033[33m'
blue = '\033[34m'
purple = '\033[35m'
cyan = '\033[36m'
white = '\033[37m'
green = '\033[32m'

# Colours Bold

red_bold = '\033[1;31m'
black_bold = '\033[1;30m'
yellow_bold = '\033[1;33m'
blue_bold = '\033[1;34m'
purple_bold = '\033[1;35m'
cyan_bold = '\033[1;36m'
white_bold = '\033[1;37m'
green_bold = '\033[1;32m'

input(prompt= green_bold+'<Press enter to Start the Game>\n')

 # room_1 gives a small introduction of story
def room_1():
    print(red_bold+f""" Heavy Lies the Crown
{'-'*20}""")

    print(black_bold+f"""
Before the Continents split, the Land was one.
It was called Kandam. Magical creatures such as Dragons 
& Yazhis(Gigantic Lions with trunks and tusks) existed here.

Of the Kingdoms in this Land, KumariKandam was the greatest kingdom.
It was ruled by the Great King Aran!.

The spy has brought some news for Aran who was in the courtyard...
{'-'*20}""")

# With below prompt, we start the game
    input(prompt= green_bold+'<Press enter to find out what the spy said. >\n')
   
    room_2() # In this room the player makes a set of decisions whether to fight or wait

def room_2():
      
    print(red_bold+f"""
    {'-'*20}    
    We are surrounded by enemy armies, my Lord!!!\n     
    """)

    #The player can either get ready for battle or consult the oracle
    decision_1= input(prompt=purple_bold+"""Quick! You can make one of the following decisions.\n
                    a. call the army general to prepare for battle, or\n
                    b. consult the oracle\n
                    
                    Choose a or b and press enter
                    
                    """)
    print(f"{'-'*20}")
    # Here the player chooses to prepare for battle.
    if decision_1.lower() == 'a' or 'call' in decision_1.lower() or 'prepare'\
    in decision_1.lower() or 'battle' in decision_1.lower() or 'army' in decision_1.lower():
        print(purple_bold+"""
        The battleground is silent. The enemies seem to be numerous.
        
                 It seems you only have two options:\n
                 a. fight or\n
                 b. surrender\n    
                 
                 """)
        #The player faces another decision on whether to fight or surrender

        decision_2= input(prompt=green_bold+"""<Choose a or b and press enter>\n""")
        print(f"{'-'*20}")  
        #Here the player chooses to fight. But it results in a fail.

        if decision_2.lower() == 'a' or 'fight' in decision_2.lower():
            print(cyan_bold+"Valiant effort. But you have been defeated.")
            fail()
            
        #Here the player chooses to surrender. But it still results in a fail.

        elif decision_2.lower()== 'b' or 'surrender' in decision_2.lower():
            print(cyan_bold+"Your Kingdom has been acquired and you were beheaded.")
            fail()

        #Adding an else statement for unknown bugs.
        else :
            print(cyan_bold+"Something went wrong. Choose wisely...")
            fail()
    #Here the player chooses to consult oracle from decision_1

    elif decision_1.lower()== 'b' or 'consult'in decision_1.lower() or 'oracle' in decision_1.lower():
        print(red_bold+f"""{'-'*20}
        Lets visit the oracle...""")
        room_3()

    #Adding an else statement for unknown bugs.      

    else :
        print(cyan_bold+"Something went wrong. Choose wisely...")
        fail()

def room_3():
    
# Oracle explains that the King must seek the snake chamber.
    print (purple_bold+"""    
    Oracle : "Death is imminent and inescapable my king. Not only from your enemies but the 
    world as we know it will be destroyed. The ground beneath us will break and the water
    around will swallow us. We must protect and pass on our knowledge to those who come next.
    Prophecy tells us that the key to achieve this is in the snake chamber of the temple. " """)
    
# The player decides to either break down the front door or inquire the priest.
    decision_3= input(prompt=purple_bold+"""Aran goes to the Temple and finds that the 
    snake chamber is sealed shut and there is no key hole. 
    You have two options:\n
    a. break down the front door or\n
    b. inquire the priest.\n    
    Choose a or b and press enter.    
    """) 
    print(f"{'-'*20}")
    sounds= ['BOOM! Hit again!', 'THUD! Hit again!', 'CRACK! What have I done?!!']
    
    # If player chooses to break, there will be 3 prompts to break the door
    
    if decision_3.lower() =='a' or 'break' in decision_3.lower() or 'down' in decision_3.lower() or 'front'\
    in decision_3.lower() or 'door' in decision_3.lower():
        
        # Loop to display sound words while breaking.
        
        for trys in sounds:
            input(prompt=yellow_bold+'< Press enter to hit the door>\n')
            print(trys)
            
        # This results in a fail scenario. We can go back to room 2.
        
        print(cyan_bold+f""" 
        {'-'*20}
        Poisonous gas is released from the hinges of the door. 
        
        Everyone in the temple is dead.
        
        Try again...
        
        """)
        input(prompt=green_bold+"<Press enter to try again>")
        room_2()
        
        #The player chooses to inquire the priest and we move to room 4.
        
    if decision_3.lower() == 'b' or 'inquire' in decision_3.lower() or 'priest' in decision_3.lower():
        
        print(red_bold+f""" 
        {'-'*20}
        Priest: It is unholy to open the snake chamber. The ancient 
        scrolls say that the we must play 3 specific notes using the 
        Rudraveena instrument to open the chamber. 
        
        Its a combination of notes C, F and B 
        
        Aran: Thats good. Cause I'm a skilled Rudraveena player. 
        """)
        
        room_4()

     # Unexpected bugs. Player can go back to room 2.
    else :
        print(f"{'-'*20}")
        print("Something went wrong. Try again...\n")
        input(prompt=green_bold+"<Press enter to try again>")
        room_2()
        
    # In this room, player is likely to win.
def room_4():
    print(black_bold+f"""
        King Aran takes his rudraveena and starts playing...

    """)
    
    # Using the random package, we pick elements from the permutations list. 
    # The seed is set so that the third attempt will result in a win scenario.
    import random
    import sys
    permutations= ["CFB","CBF","BCF","BFC","FCB","FBC"]
    random.seed(7)
    triads=0
    # Adding a loop for allowing only three attempts.
    while triads<3:
        attempt= input(prompt=green_bold+"<Press enter to play the instrument>")
        notes=random.choice(permutations)
        print(blue_bold+f"You played {notes}. That's a nice tune!")
        triads+=1    
        if notes=='BFC':
            
            # Some conclusion to the win scenario.
            
            print(yellow_bold+f"""
            {'-'*20}
The hinges of the door snap out. The door falls backward. Thud!!!
A huge golden Yazhi stares right at Aran. Not to worry. It is a statue.
Below the statue were a stack of ancient scrolls.They all had an inscription of a Python.                  
       
Aran went through the scrolls and found a Teleportation mantra.                  
He read aloud "Far into the future, Control my travel, Be it to a time, where everything is mature."

The eyes of the Yazhi were suddenly radiant with gleaming light. The priest shielded his eyes. Boom!
He opened his eyes...he smiled. The Yazhi and the King were gone. And so were the ancient scrolls.
                                                     
                                        You win!                     
                  """)
            input(green_bold+'<Press enter to exit>')
            sys.exit(0)
            
            # For the first two attempts the below statement is executed.
            
        elif notes != 'BFC':
            print(green_bold+f"""
            
            Oops. Wrong notes. Try again.
            {'-'*20}""")
            
            #In case of unexpected bugs we send the player back to room 3.
        
        else:
            print(green_bold+"Something went wrong. Try again...")
            
            input(green_bold+'<Press enter to try again>')
            room_3()
        
        

def fail():
    import sys
    print(f"{'-'*20}")
    print(cyan_bold+"\nYou've failed, but thanks for playing!")
    input(green_bold+'<Press enter to exit>\n')
    sys.exit(0)        

room_1()        


```
